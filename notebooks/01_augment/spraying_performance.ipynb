```python
# Задаём пути и читаем исходные данные
root = "../.."
path_from = f"{root}/data/raw"
path_to = f"{root}/data/augmented"
name = "spraying_performance"

data = pd.read_csv(f"{path_from}/{name}.csv")
```

```python
# Настраиваем, какие поля и на сколько варьировать
schema = {
    "experiment.weather.temperature.min": (0.02, 0.05),
    "experiment.weather.temperature.max": (0.02, 0.05),
    "experiment.weather.humidity.min": (0.02, 0.05),
    "experiment.weather.humidity.max": (0.02, 0.05),
    "experiment.weather.velocity.min": (0.05, 0.1),
    "experiment.weather.velocity.max": (0.05, 0.1),
    "experiment.params.height": (0.05, 0.1),
    "experiment.params.flight_speed": (0.05, 0.1),
    "experiment.params.spray_volume": (0.05, 0.1),
    "experiment.results.coverage.value": (0.01, 0.03),
    "experiment.results.coverage.max": (0.01, 0.03),
    "experiment.results.coverage.CV": (0.01, 0.03),
    "experiment.results.droplet_size.value": (0.01, 0.03),
    "experiment.results.droplet_size.max": (0.01, 0.03),
    "experiment.results.droplet_size.CV": (0.01, 0.03),
}
```

```python
# Аугментация ×5 и сохранение
data_augmented_x5 = augment_dataframe(data, schema)
data_augmented_x5 = fix_max_ge_value(data_augmented_x5, schema)
data_augmented_x5.to_csv(f"{path_to}/{name}_x5.csv", index=False)
```

```python
# Аугментация ×10 и сохранение
data_augmented_x10 = augment_dataframe(data, schema, n=9)
data_augmented_x10 = fix_max_ge_value(data_augmented_x10, schema)
data_augmented_x10.to_csv(f"{path_to}/{name}_x10.csv", index=False)
```

```python
# Порядок экспериментов для T30 и T40
experiment_order_T30 = [
    "H1-S2-V1", 
    "H1-S1-V1", 
    "H2-S1-V1", 
    "H3-S1-V1", 
    "H1-S1-V2", 
    "H2-S1-V2", 
    "H3-S1-V2", 
    "H1-S1-V3", 
    "H2-S1-V3", 
    "H3-S1-V3"
]

experiment_order_T40 = [
    "H1-S1-V1-D2",
    "H2-S1-V1-D2",
    "H3-S1-V1-D2",
    "H1-S1-V2-D2",
    "H2-S1-V2-D2",
    "H3-S1-V2-D2",
    "H1-S1-V3-D2",
    "H2-S1-V3-D2",
    "H3-S1-V3-D2",
    "H1-S2-V2-D2",
    "H2-S2-V2-D2",
    "H3-S2-V2-D2",
    "H2-S1-V2-D3",
    "H2-S1-V2-D1"
]
```

```python
# Coverage для AGRAS T30
plot_metric_bw_style(
    data, data_augmented_x5, data_augmented_x10,
    "AGRAS T30",
    "coverage",
    experiment_order=experiment_order_T30,
    y_value_range=(0, 16),
    y_cv_range=(0, 80)
)
```

```python
# Droplet Size для AGRAS T30
plot_metric_bw_style(
    data, data_augmented_x5, data_augmented_x10,
    "AGRAS T30",
    "droplet_size",
    experiment_order=experiment_order_T30,
    y_value_range=(0, 360),
    y_cv_range=(0, 40)
)
```

```python
# Coverage для AGRAS T40
plot_metric_bw_style(
    data, data_augmented_x5, data_augmented_x10,
    "AGRAS T40",
    "coverage",
    experiment_order=experiment_order_T40,
    y_value_range=(0, 18),
    y_cv_range=(0, 120)
)
```

```python
# Droplet Size для AGRAS T40
plot_metric_bw_style(
    data, data_augmented_x5, data_augmented_x10,
    "AGRAS T40",
    "droplet_size",
    experiment_order=experiment_order_T40,
    y_value_range=(0, 600),
    y_cv_range=(0, 50)
)
```

При внимательном рассмотрении каждого кластера столбиков, видно: ни пятикратная, ни десятикратная аугментация не вносят существенных смещений ни в сами значения метрик, ни в их разброс (CV). Так, для T30–coverage средние высоты столбиков «aug ×5» и «aug ×10» практически совпадают с исходным: отклонения измеряются сотыми или десятыми долями единицы, что находится в пределах расчетной погрешности. Линии CV для всех трёх наборов данных идут практически вровень — никакой дополнительной нестабильности или резкого изменения дисперсии после аугментации не появляется.

Аналогичная картина просматривается и у T30–droplet_size, и у обеих метрик в блоках T40. Везде разница между «orig», «aug ×5» и «aug ×10» не превышает нескольких процентов от среднего, а линии CV лежат почти друг на друге. Поэтому можно заключить: используемая аугментация не искажает результаты за пределы естественного статистического разброса и не вносит систематической погрешности в оценку покрываемости или размера капель.
