```python
# Задаём пути и читаем исходные данные
root = "../.."
path_from = f"{root}/data/raw"
path_to = f"{root}/data/augmented"
name = "droplet_deposition"

data = pd.read_csv(f"{path_from}/{name}.csv")
```

```python
# Настраиваем, какие поля и на сколько варьировать
schema = {
    "experiment.weather.temperature": (0.02, 0.05),
    "experiment.weather.humidity": (0.02, 0.05),
    "experiment.weather.velocity": (0.05, 0.1),
    "experiment.params.work_height": (0.05, 0.1),
    "experiment.params.height": (0.05, 0.1),
    "experiment.params.flight_speed": (0.05, 0.1),
    "experiment.results.coverage.value": (0.01, 0.03),
    "experiment.results.coverage.CV": (0.01, 0.03),
    "experiment.results.droplet_size.value": (0.01, 0.03),
    "experiment.results.droplet_size.CV": (0.01, 0.03),
}
```

```python
# Аугментация ×5 и сохранение
data_augmented_x5 = augment_dataframe(data, schema)
data_augmented_x5.to_csv(f"{path_to}/{name}_x5.csv", index=False)
```

```python
# Аугментация ×10 и сохранение
data_augmented_x10 = augment_dataframe(data, schema, n=9)
data_augmented_x10.to_csv(f"{path_to}/{name}_x10.csv", index=False)
```

```python
# Порядок экспериментов для T20-coverage и T20-droplet-size
experiment_order_coverage = [f"T{i}" for i in range(1,10)]
experiment_order_droplet_size = [f"T{i}" for i in range(1,4)]
```

```python
# Coverage для AGRAS T20
plot_metric_bw_style(
    data, data_augmented_x5, data_augmented_x10,
    "AGRAS T20",
    "coverage",
    experiment_order=experiment_order_coverage,
    y_value_range=(0, 10),
    y_cv_range=(0, 80)
)
```

```python
# Droplet Size для AGRAS T20
plot_metric_bw_style(
    data, data_augmented_x5, data_augmented_x10,
    "AGRAS T20",
    "droplet_size",
    experiment_order=experiment_order_droplet_size,
    y_value_range=(0, 450),
    y_cv_range=(0, 20)
)
```

На графике «AGRAS T20 — coverage» столбики «orig», «aug ×5» и «aug ×10» идут практически вплотную друг к другу: для каждого из трёх опытов (T1, T2, T3 и т. д.) разница в средней площади покрытия измеряется сотыми долями единицы. Это означает, что добавление синтетических образцов ни при пятикратном, ни при десятикратном увеличении не вносит заметного смещения в результат измерения покрытия. Линия CV (коэффициента вариации) также практически совпадает для всех трёх наборов данных — разброс значений остаётся стабильным и лежит в окрестности 40–55 %, что укладывается в естественные границы статистической погрешности.

На втором графике «AGRAS T20 — droplet_size» та же картина: высоты столбцов «droplet_size value» для оригинала и аугментаций отличаются лишь на десятки единиц (менее 1 % от среднего порядка 300–360 мкм). А линии CV располагаются так плотно, что визуально сливаются в одну кривую со значениями около 10–12 %. Следовательно, при работе с T20 аугментация данных в пяти- и десятикратном вариантах не оказывает существенного влияния ни на средний размер капель, ни на единообразие их распределения.
